### Extracting data from different endpoints of the API
Now we have the base dataframe. Follwing helper functions will extract useful data from the base dataframe and store them in these List variables.

```python
#Lists for storing extracted data from the base dataset 
BoosterVersion = []
PayloadMass = []
Orbit = []
LaunchSite = []
Outcome = []
Flights = []
GridFins = []
Reused = []
Legs = []
LandingPad = []
Block = []
ReusedCount = []
Serial = []
Longitude = []
Latitude = []
```

**getBoosterVersion()**: <code>rocket</code> column of the base dataframe provide IDs for different boosters/rockets. This function will extract Booster Version names from the API.

```python
def getBoosterVersion(data):
    """The helper function for extract data from SpaceX API.
    
    parameters: data - base dataframe
    output: 
    - Booster versions"""
    
    for x in data['rocket']:
        if x:
            response = requests.get("https://api.spacexdata.com/v4/rockets/"+str(x)).json()
            BoosterVersion.append(response['name'])
```

**getLaunchSite()**: <code>launchpad</code> column of the base dataframe provides IDs for different launchpads. This function extracts
- Launch site name
- Latitude and Longitude

```python
def getLaunchSite(data):
    """The helper function for extract data from SpaceX API.
    
    parameters: data - base dataframe
    output: 
    - Latitude of launch sites
    - Longitude of launch sites
    - Launch site names"""
            
    for x in data['launchpad']:
        if x:
            response = requests.get("https://api.spacexdata.com/v4/launchpads/"+str(x)).json()
            Longitude.append(response['longitude'])
            Latitude.append(response['latitude'])
            LaunchSite.append(response['name'])
```

**getPayloadData()**: <code>payloads</code> column of the base dataset provides IDs for different payloads. This function extracts
- Payload Mass
- Orbit

```python
def getPayloadData(data):
    """The helper function for extract data from SpaceX API.
    
    parameters: data - base dataframe
    output: 
    - Payload Mass
    - Orbit"""
 
    for load in data['payloads']:
        if load:
            response = requests.get("https://api.spacexdata.com/v4/payloads/"+load).json()
            PayloadMass.append(response['mass_kg'])
            Orbit.append(response['orbit'])
```

**getCoreData()**: From <code>cores</code> we would like to learn the outcome of the landing, the type of the landing, number of flights with that core, whether gridfins were used, whether the core is reused, whether legs were used, the landing pad used, the block of the core which is a number used to seperate version of cores, the number of times this specific core has been reused, and the serial of the core.

```python
def getCoreData(data):
    """The helper function for extract data from SpaceX API.
    
    parameters: data - base dataframe
    output: 
    - Outcome and type of landing
    - Number of flights with that core
    - Grifins used/not
    - core was reused/not
    - Legs were used/not
    - Landingpad was used/not"""
    
    for core in data['cores']:
        if core['core'] is not None:
            response = requests.get("https://api.spacexdata.com/v4/cores/"+core['core']).json()
            Block.append(response['block'])
            ReusedCount.append(response['reuse_count'])
            Serial.append(response['serial'])
        else:
            Block.append(None)
            ReusedCount.append(None)
            Serial.append(None)
        Outcome.append(str(core['landing_success'])+' '+str(core['landing_type']))
        Flights.append(core['flight'])
        GridFins.append(core['gridfins'])
        Reused.append(core['reused'])
        Legs.append(core['legs'])
        LandingPad.append(core['landpad'])
```

Arrange the base dataframe removing unwanted columns.

```python
# Lets take a subset of our dataframe keeping only the features we want and the flight number, and date_utc.
sx_df_base = sx_df_base[['rocket', 'payloads', 'launchpad', 'cores', 'flight_number', 'date_utc']]

# We will remove rows with multiple cores because those are falcon rockets with 2 extra rocket boosters and rows that have multiple payloads in a single rocket.
sx_df_base = sx_df_base[sx_df_base['cores'].map(len) == 1]
sx_df_base = sx_df_base[sx_df_base['payloads'].map(len) == 1]

# Since payloads and cores are lists of size 1 we will also extract the single value in the list and replace the feature.
sx_df_base['cores'] = sx_df_base['cores'].map(lambda x: x[0])
sx_df_base['payloads'] = sx_df_base['payloads'].map(lambda x: x[0])

# We also want to convert the date_utc to a datetime datatype and then extracting the date leaving the time
sx_df_base['date'] = pd.to_datetime(sx_df_base['date_utc']).dt.date
```

#### Construct final dataframe

```python
launch_dict = {
    'FlightNumber': list(sx_df_base['flight_number']),
    'Date': list(sx_df_base['date']),
    'BoosterVersion': BoosterVersion,
    'PayloadMass': PayloadMass,
    'Orbit': Orbit,
    'LaunchSite': LaunchSite,
    'Outcome': Outcome,
    'Flights': Flights,
    'GridFins': GridFins,
    'Reused': Reused,
    'Legs': Legs,
    'LandingPad': LandingPad,
    'Block': Block,
    'ReusedCount': ReusedCount,
    'Serial': Serial,
    'Longitude': Longitude,
    'Latitude': Latitude}
```

```python
# create dataframe from launch_dict dictionary
data_falcon = pd.DataFrame(launch_dict)
```

### Simple data cleaning process
There are some missing values in *PayloadMass* column. *PayloadMass* is a continueous feature. There are 22 missing values. Considering the size of the dataset, we cannot remove those observations. So, in this case I use ***mean*** value of the column to fill in those missing values.

```python
data_falcon9.isna().sum()
```

```python
mean_of_payloads = data_falcon9['PayloadMass'].mean()
data_falcon9['PayloadMass'] = data_falcon9['PayloadMass'].fillna(mean_of_payloads)
```

```python
data_falcon9.isna().sum()
```

```python
data_falcon9[data_falcon9['Orbit'].isna()]
```

I searched Falcon 9 Wikipedia page and found out that the missing **Orbit** is actually **LEO**



So, I replaced missing orbit with LEO.

```python
data_falcon9['Orbit'].fillna('LEO', inplace=True)
```

```python
data_falcon9.isna().sum()
```
